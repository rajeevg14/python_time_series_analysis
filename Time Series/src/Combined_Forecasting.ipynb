## Customer Demand Forecasting

```python
import pandas as pd
hours=pd.read_csv('../data/input/hour.csv')
```

```python
hours.head()
```

```python
hours.tail()
```

## Plotting Data to Find Trends

```python
from matplotlib import pyplot as plt
plt.scatter(hours['dteday'],hours['cnt'])
plt.title('Customer Demand by Hour')
plt.xlabel('Hour')
plt.ylabel('Demand')
plt.show()
```

## Plotting Line Graph

```python
fig, ax = plt.subplots(figsize=(10, 6))
#ax.plot() draws the plot
ax.plot(hours['dteday'], hours['cnt'],c='red',label='total',linestyle='-')
ax.legend()
plt.show()
```
